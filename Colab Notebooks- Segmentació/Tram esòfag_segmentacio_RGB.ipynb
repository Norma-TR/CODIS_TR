# 2. Funció per segmentar la imatge RGB  

La funció té com a entrada la imatge RGB I i retorna la imatge segmentada L

```python
import numpy as np
from sklearn.cluster import KMeans

X = np.reshape(I,(I.shape[0]*I.shape[1],I.shape[2]))
labels = KMeans(n_clusters=4).fit_predict(X)
L = np.reshape(labels,(I.shape[0],I.shape[1]))
```

# 3. grafiquem la imatge original i el resultat de la segmentació

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(121)
ax1 = plt.imshow(I)

ax2 = fig.add_subplot(122)
ax2 = plt.imshow(L)
```
