Ara farem un bucle que imprimeixi tots els números entre 0 i 20

```python
for k in range(21):
  print(k)
```

Fem un programa que sumi els valors del 0 al 20:

```python
suma = 0
for k in range(21):
  suma = suma + k
print(suma)
```

Si vull sumar els numeros entre 1.000.000 i 2.000.000 llavors farem que la instrucció range pregui tres paràmetres: el primer és el valor inicial, el segon és el valor final i el tercer és si volem anar d'un en un o de dos en dos o tres en tres, etc

```python
suma = 0
for k in range(1000000,2000000,1):
  suma = suma + k
print(suma)
```
