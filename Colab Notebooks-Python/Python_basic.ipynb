## Variables

Tots els programes de programació treballen amb variables. Són tipus de dades que permeten emmagatzemar i manipular infomació. Per exemple, podem definir una variable anoemanda 'x' i assignar-li un valor de 3.14 de la forma següent

```python
x = 3.14 # aswsignem el valor 3.14 a la variable x
```

podem consultar el valor de la variable x fent això

```python
x
```

També pots fer que s'imprimeixi un missatge en el que estigui la variable x, per exemple:

```python
print('El valor de la variable x és: {}'.format(x))
```

Les variables poden ser numèriques i també cadenes de caràcters alfanumèrics, per exemple:

```python
a = 'Hola que tal'
print(a)
```

Podem fer operacions entre variables, per exemple

```python
x = 25
y = 5
z = x*y
print('El resultat de {} x {} és {}'.format(x,y,z))
```

## Llibreries i funcions

Python és un llenguatge que funciona amb llibreries i funcions que ja estan fetes i que pots fer servir pel que necessitis. Per fer servir una funció d'una llibreria ho hem d'especificiar al codi, per exemple, volem trobar el resultat de l'operació $3*\cos(2*\pi)$

```python
import math # el primer es importar la llibreria que farem servir

x = 3*math.cos(2*math.pi) # de la llibreria math fem servir les funcions cos i pi (constant)
print(x)
```

També es poden importar només les funcions que farem servir, per exemple:

```python
from math import cos, pi

x = 3*cos(2*pi)
print(x)
```

El més habitual és fer servir un nickname per referir-nos a la llibreria

```python
import math as m

x = 3*m.cos(2*m.pi)
print(x)
```

## Bucles i loops

Sovint necessitem que una variable vagi canviant de valors seguint un ordre. Per aixó farem servir la funció loop, per exemple



```python
for i in range(0,10,1): # lavariable x iterara entre 0 i 10 d'un en un
    print(i) # important indentar la linia per a que estigui dintre del loop
```

Si no indentem no funciona

```python
for i in range(0,10,1):
print(i)
```

Podem fer un loop amb una variable que no sigui numèrica, per exemple

```python
vector = ["hola","buenos","dias"]
for palabra in vector:
    print (palabra)
```

##Condicionals

De vegades el porgrama necessita fer una cosa quan una variable prengui un valor i un altre quan prengui un altre, per exemple

```python
for i in range(0,10,1):

  if i>5:
      print(i)
  else:
      print('no es mes gran que 5')
```
