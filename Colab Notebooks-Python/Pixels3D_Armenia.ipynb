#3. Representar els pixels com una diagrama de dispersió 3D a l'espai RGB

Seguim l'exemple de la funció scatter3D de la llibreria matplotlib
https://matplotlib.org/stable/gallery/mplot3d/scatter3d.html


```python
# Redimensionem la imatge com una matriu de dimensions (NPIX,3)

import numpy as np

X = np.reshape(I,(I.shape[0]*I.shape[1],I.shape[2]))
X.shape
```

```python
I[0,0] # el píxel de dalt a l'esquerra
```

```python
I[I.shape[-1],0] # el píxel de baix l'esquerra
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(projection='3d')

xred = X[:,0] # valor de tots els pixels al canal red
xgreen = X[:,1] # valor de tots els pixels al canal green
xblue = X[:,2] # valor de tots els pixels al canal blue

ax.scatter(xred, xgreen, xblue, marker='.')

ax.set_xlabel('RED')
ax.set_ylabel('GREEN')
ax.set_zlabel('BLUE')

plt.show()
```
